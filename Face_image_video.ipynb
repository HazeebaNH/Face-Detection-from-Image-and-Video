```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('data/group.jpg')
```

```python
face_cascade=cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')                     #to read haarcascade file
```

```python
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_img)                                #1:02:00
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray)                                            #don't mind the yellowish tone of grayscale converted image
```

```python
faces=face_cascade.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))
                                           #img_gray - face is detected from grayscale image and not from RGB image
                                           #Rest 3 parameters(scaleFactor,minNeighbors,minSize) have fixed or default values always.
                                           #These 3 parameters are meant to automatically detect from values of nearby pixels
                                           #scaleFactor - scale of pixels
                                           
```

```python
faces           #o/p represents the detected 8 faces in image and their respective 4 coordinates (say, x,y,w,h) 
                #(actually, there are only 6 faces in the image)
                
```

```python
len(faces)                                 #another command to get no. of faces detected, when there are many faces or images
```

```python
#here, we draw rectangles for each faces using "for loop"
for (x,y,w,h) in faces:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),(0,255,0),7)                #(0,255,0) - green colour    
                                           
          #among coordinates in o/p array,
              #x and y are points. w and h are width and height resp. But when we drew rectangle in earlier file "introduction1.ipynb", we passed 4 points of image.
              #So, inorder to convert w and h into points, we need to give 3rd and 4th coordinates as x+w and y+h resp to draw rectangle. 
```

```python
#to detect face in video
webcam=cv2.VideoCapture('/home/hazeeba/Videos/DL_session/face_detection/data/face video.mp4')             #give the path of video in bracket


#or  webcam=cv2.VideoCapture(0)    # - this code gives access to our webcam
             #webcam opens when any one among the three values (0 or 1 or -1) is given in above code
             #after running this code,run the below cell with "while" loop.Then webcam opens up in window and closes when 'esc' key is pressed
             #the value among the 3, for which our webcam opens, is the index of our webcam
```

```python
while True:
    success,frame=webcam.read()            #2 parameters to read webcam: (1)success - to check whether camera is success or not,ie, whether camera produces o/p or not
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)                       #  (2)frame   - the o/p of webcam is stored in frame

 
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))            
    print(len(faces))
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),5)                                  

    cv2.imshow('face detection',frame)           #to open webcam or to view frame
    if cv2.waitKey(1) & 0xFF==27:                #0xFF - zero ex ff
        break                                    #this if loop is meant to stop window automatically after the video is completely run or when the esc key is pressed
webcam.release()
cv2.destroyAllWindows()               #when this cell is run, video plays in high speed if codes b/w success and cv2.imshow are deleted
                                      #while running this cell,press 'esc' key to exit and don't wait for the video to run completely, as this will result
                                         #in breakdown of kernel and then we need to restart kernel and run all cells from beginning       
                                     #this issue happens only with video and no issue with webcam  
                                     
                                      
                                      
                                     #when this entire cell is run,face detection is shown in video using rectangle     
                                     #each time before running this cell, the cell above must be run to get o/p. Or else, type the code in 
                                     #above cell in this cell itself.                      
```
